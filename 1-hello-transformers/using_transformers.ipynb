```python
from PIL import Image
import requests
from transformers import pipeline
import numpy as np
from io import BytesIO
from skimage import io
from skimage import color
import matplotlib.pyplot as plt

url = "https://www.wtso.com/blog/wp-content/uploads/2018/07/shutterstock_492006757.jpg"

response = requests.get(url)
image = Image.open(BytesIO(response.content))

image_segmenter = pipeline("image-segmentation")
seg_maps = image_segmenter(image)

image = np.array(image)
masks = [np.array(s['mask']) for s in seg_maps]

seg_mask = np.zeros(image.shape[0:2])

for id, m in enumerate(masks):
  m = np.array(m)
  m_temp = m
  m_temp[m_temp != 0] = id + 1
  seg_mask = seg_mask + m_temp

io.imshow(color.label2rgb(seg_mask, image))

plt.show()
```
